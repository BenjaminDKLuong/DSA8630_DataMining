```r
glass_data[["refractive_index"]] <- ordered(cut(glass_data[["refractive_index"]], c(1.51,1.517,1.518,1.519,1.534)), labels = c("low", "medium", "high", "very high"))
glass_data[["Sodium"]] <- ordered(cut(glass_data[["Sodium"]], c(10.72, 12.91, 13.30, 13.82, 17.38)), labels = c("low", "medium", "high", "very high"))
glass_data[["Magnesium"]] <- ordered(cut(glass_data[["Magnesium"]], c(-1,2.115,3.48,3.6,4.49)), labels = c("low", "medium", "high", "very high"))
glass_data[["Aluminum"]] <- ordered(cut(glass_data[["Aluminum"]], c(0,1.19,1.36,1.63,3.5)), labels = c("low", "medium", "high", "very high"))
glass_data[["Silicon"]] <- ordered(cut(glass_data[["Silicon"]], c(69.8,72.28,72.79,73.09,75.41)), labels = c("low", "medium", "high", "very high"))
glass_data[["Potassium"]] <- ordered(cut(glass_data[["Potassium"]], c(-1,0.1225,0.555,0.61,6.21)), labels = c("low", "medium", "high", "very high"))
glass_data[["Calcium"]] <- ordered(cut(glass_data[["Calcium"]], c(5.4,8.24,8.6,9.172,16.190)), labels = c("low", "medium", "high", "very high"))
glass_data[["Barium"]] <- ordered(cut(glass_data[["Barium"]], c(-1, 0 , 1.05, 2.1, 3.15)), labels = c("low", "medium", "high", "very high"))
glass_data[["Iron"]] <- ordered(cut(glass_data[["Iron"]], c(-1,0,0.17,0.34,0.51)), labels = c("low", "medium", "high", "very high"))
glass_data[["Type_of_glass"]] <-factor(glass_data[["Type_of_glass"]], c(1,2,3,5,6,7), labels = c("building_windows_float_processed", "building_windows_non_float_processed", "vehicle_windows_float_processed","containers","tableware","headlamps"))
```

**Activity 3:** Now, recode the data as a binary incidence matrix by coercing the data set to transactions. Save the transactions to glass_trans variable.

```r
library("arules")
glass_trans <- as(glass_data, "transactions")
```

**Activity 4:** Draw a frequency plot for the generated transactions with a support of 0.1

```r
## Your code for activity 4 goes here
## -----------------------------------
itemFrequencyPlot(glass_trans, support = 0.1, cex.names=0.8)
```

**Activity 5:** Display the first three rules using inspect function. Use a support of 0.1 and a confidence of 0.6.

```r
## Your code for activity 5 goes here
## -----------------------------------
rules <- apriori(glass_trans, parameter = list(support = 0.1, confidence = 0.6))
inspect(head(rules,n=3))
```

**Activity 6:** Draw a scatter plot for the generated rules using measures 'support', 'lift' and use 'confidence' for shading.

```r
## Your code for activity 6 goes here
## -----------------------------------
library("arulesViz")
plot(rules, measure=c("support", "lift"), shading="confidence")
```

**Activity 7:** Draw a matrix plot for the generated rules using the measure 'lift'. Order the transactions by the measure "lift".

```r
plot(rules, method="matrix", measure=c("lift"), control=list(reorder="measure"))
```

**Activity 8:** Draw a graph for the top 10 generated rules.

```r
plot(head(rules,10), method="graph")
```
