**Activity 1: ** Scrap the wiki page "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_sector_composition" to extract the content. Create a soup object using Beautiful soup library and save the soup in a variable called wiki_soup  

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_sector_composition"
response = requests.get(url)
wiki_soup = BeautifulSoup(response.text, "html")
```

**Activity 2: ** Extract the table "GDP from natural resources" from the soup and print it.

```python
all_tables = wiki_soup.find_all('table')
len(all_tables)
```

```python
# the table we are looking for
table = all_tables[4]
#table
```

**Activity 3: ** Create a dataframe called "resources_df" from the extracted table. Use the column names same as the column headings in original wiki table but make them valid. For example, use "country" instead of "country/economy". 

```python
Country = []
Total_natural_resources = []
Oil = []
Natural_gas = []
Coal = []
Mineral = []
Forest = []

for row in table.findAll("tr")[1:-1]:
    columns = row.findAll('td')
    if columns[0].find('a'):
        Country.append(columns[0].find('a').find(text=True))
    else:
        Country.append(columns[0].find(text=True))
    
    Total_natural_resources.append(columns[1].find(text=True))
    Oil.append(columns[2].find(text=True))
    Natural_gas.append(columns[3].find(text=True))
    Coal.append(columns[4].find(text=True))
    Mineral.append(columns[5].find(text=True))
    Forest.append(columns[6].find(text=True))

import pandas as pd
resources_df=pd.DataFrame(Country,columns=['Country'])
resources_df['Total_natural_resources']=Total_natural_resources
resources_df['Oil']=Oil
resources_df['Natural_gas']=Natural_gas
resources_df['Coal']=Coal
resources_df['Mineral']=Mineral
resources_df['Forest']=Forest
resources_df.head()
```

**Activity 4: ** Extract the table "gdp per person employed(ppp) (2015) by sector" from wiki page and create a dataframe called "gdp_percent" out of it. Use the column names same as the column headings in original wiki table but make them valid. For example, use "country" instead of "country/economy". 

```python
# the table we are looking for
table2 = all_tables[6]
table2
```

```python
Country = []
Agriculture_of_GDP = []
Industry_of_GDP = []
Services_of_GDP = []
Agriculture_of_employment = []
Industry_of_employment = []
Services_of_employment = []

for row in table2.findAll("tr")[1:]:
    columns = row.findAll('td')
    if columns[0].find('a'):
        Country.append(columns[0].find('a').find(text=True))
    else:
        Country.append(columns[0].find(text=True))
    Agriculture_of_GDP.append(columns[1].find(text=True))
    Industry_of_GDP.append(columns[2].find(text=True))
    Services_of_GDP.append(columns[3].find(text=True))
    Agriculture_of_employment.append(columns[4].find(text=True))
    Industry_of_employment.append(columns[5].find(text=True))
    Services_of_employment.append(columns[6].find(text=True))


import pandas as pd
gdp_percent=pd.DataFrame(Country,columns=['Country'])
gdp_percent['Agriculture_of_GDP']=Agriculture_of_GDP
gdp_percent['Industry_of_GDP']=Industry_of_GDP
gdp_percent['Services_of_GDP']=Services_of_GDP
gdp_percent['Agriculture_of_employment']=Agriculture_of_employment
gdp_percent['Industry_of_employment']=Industry_of_employment
gdp_percent['Services_of_employment']=Services_of_employment
gdp_percent.head()
```

**Activity 5: ** Combine the dataframes resources_df and gdp_percent. Name the resultant dataframe as combined_df

```python
combined_df = pd.merge(resources_df,gdp_percent,how='inner',on='Country')
combined_df.head()
```

**Activity 6: ** Replace the the invalid values ".." with valid NAN in combined_df

```python
combined_df.replace('..', np.NaN, inplace = True)
combined_df.head()
```
